# IV. Data Cleaning

11. (3б) Перетворити колонку launched у цілочислену (Int64). Перевірити, чи все пройшло, як очікувалось.

```python
df_missile_attacks["launched"] = df_missile_attacks["launched"].astype("Int64")
```

```python
df_missile_attacks["launched"].dtype
```

12. (3б) Видалити колонки, де кількість непустих значень становить менше 30% від загальної кількості записів. Можна вручну порахувати і тут тільки видалити. За автоматизацію розрахунку відстоків - додаткові 3 бали.

```python
threshold = len(df_missile_attacks) * 0.3
to_delete = [col for col in df_missile_attacks.columns if df_missile_attacks[col].count() < threshold]
result = df_missile_attacks.drop(columns=to_delete)
result
```

13. (4б) Очистити датафрейм від дублікатів (дублюючихся рядків) та записів, де атака продовжувалась відʼємну або нульову кількість часу (що фізично неможливо). Не забудьте оновити індекс. Скільки рядків було видалено? Як ви це перевірили у коді?

```python
df_missile_attacks['time_start'] = pd.to_datetime(df_missile_attacks['time_start'], errors='coerce')
df_missile_attacks['time_end'] = pd.to_datetime(df_missile_attacks['time_end'], errors='coerce')

df_missile_attacks['duration'] = (df_missile_attacks['time_end'] - df_missile_attacks['time_start']).dt.total_seconds() / 60

init_count = len(df_missile_attacks)
print(init_count)

filtered_missile_attacks = df_missile_attacks[df_missile_attacks['duration'] > 0]

filtered_missile_attacks = filtered_missile_attacks.drop_duplicates().reset_index(drop=True)

filtered_missile_attacks
```

```python
final_count = len(filtered_missile_attacks)

removed = init_count - final_count

removed
```

16. (11б) Зʼясуйте 10 найпопулярніших моделей запущених пристроїв за кількістю випущених одиниць за наведений період. Запишіть результат у файл типу json. Скопіюйте зміст файлу та вставте нижче.

17. (10б) Зʼясуйте, у який день було найбільше влучань за весь період (незбитих пристроїв). Виведіть час початку атаки та кількість влучань.

```python
df_missile_attacks['hits'] = df_missile_attacks['launched'] - df_missile_attacks['destroyed']
most_hits_date = df_missile_attacks.groupby(df_missile_attacks['time_start'].dt.date)['hits'].sum().idxmax()
most_hits_count = df_missile_attacks.groupby(df_missile_attacks['time_start'].dt.date)['hits'].sum().max()

(most_hits_date, most_hits_count)
```

18. (15б) Знайдіть найбільш тривалу та найбільш коротку атаки за кожен представлений місяць. Результат отримайте у такому вигляді:

![текст посилання](https://drive.google.com/uc?id=1gISSxR2mXjTEOgoAk_vQnXJeqoL-eA5b)


Запишіть його у файл типу сsv. Після цього виведіть на екран.

```python
df_missile_attacks['year'] = df_missile_attacks['time_start'].dt.year
df_missile_attacks['month'] = df_missile_attacks['time_start'].dt.month

df_missile_attacks['time_start'] = pd.to_datetime(df_missile_attacks['time_start'])
df_missile_attacks['time_end'] = pd.to_datetime(df_missile_attacks['time_end'])

df_missile_attacks['attack_duration'] = df_missile_attacks['time_end'] - df_missile_attacks['time_start']

shortest = df_missile_attacks.loc[df_missile_attacks.groupby(['year', 'month'])['attack_duration'].idxmin()]
longest = df_missile_attacks.loc[df_missile_attacks.groupby(['year', 'month'])['attack_duration'].idxmax()]

result = pd.concat([longest, shortest]).sort_values(by=['year', 'month', 'attack_duration'])

result[['year', 'month', 'attack_duration']].to_csv('attack_durations_by_month.csv', index=False)

print(result[['year', 'month', 'attack_duration']])
```
